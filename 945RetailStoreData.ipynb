```python
# New project to import Retail data into local machine after cleaning the data

import pandas as pd
import numpy as np


data = pd.read_csv(f"https://drive.google.com/uc?id=15akNjGB2rSjQwz7sXsWevZk0xDqarNja", encoding="utf-8")

print(data.head(6))


print('Data size : ', data.shape)

print('Null values per column : \n', data.isnull().sum())

data = data.dropna()


print(data.describe())

# Save processed data to an Excel file with a tab named "Raw Data"
output_file = "RetailStore.xlsx"  # Specify the file name
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    data.to_excel(writer, sheet_name="Raw Data", index=False)

print(f"Processed data has been saved to {output_file} in the 'Raw Data' tab.")
```
