```python
# Importing Telco Customer Churn dataset from Github and storing in a local SQL table

import pandas as pd
import sqlite3

url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"
df = pd.read_csv(url)
print("Data imported successfully. Here's a quick preview:")
print(df.head())


conn = sqlite3.connect("customer_data.db")
df.to_sql("Telco_customers", conn, if_exists="replace", index=False)
print("\nData written to SQL table 'customers' successfully.")


query = "SELECT * FROM Telco_customers LIMIT 10"
preview_df = pd.read_sql_query(query, conn)
print("\nPreview of the first 10 rows from the 'customers' table:")
print(preview_df)

conn.close()
```
