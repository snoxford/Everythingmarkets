```python
# This program will take Product Sales data from local drive# will clean it up for blank spaces, NaNs and non-logical data# will update the input file with Summary Sales and Average Prices by productimport pandas as pd
import numpy as np

df = pd.read_excel('Products_data.xlsx',header = 0)

df.dropna(how='all', inplace=True)

df.dropna(axis=1, how='all', inplace=True)

df.columns = df.iloc[0]
df = df.drop(0).reset_index(drop=True)

df.replace('', np.nan, inplace=True) 
df.dropna(inplace=True) 
df = df[~(df == 0).any(axis=1)]

# Wherever Price is non-numeric - > change it to NaN
df['Price (USD)'] = pd.to_numeric(df['Price (USD)'], errors='coerce') 
df.dropna(subset=['Price (USD)'], inplace=True)

# Sort by brand and reset the index
df = df.sort_values(by='Brand')
df.reset_index(drop=True, inplace=True)

print(df)


#Give the Sales by Brand
Brand_counts = df.groupby('Brand').size().reset_index(name='Total Sales')
print('Sales by Brand:',Brand_counts)

#Give average price for the Brands
Brand_avg_price = df.groupby('Brand')['Price (USD)'].mean().reset_index(name='Average_price') 
print('Average Price by Brand:',Brand_avg_price)

with pd.ExcelWriter('Products_data.xlsx', mode='a', engine='openpyxl') as writer: 
    Brand_counts.to_excel(writer, sheet_name='Brand_Total_Rows', index=False) 
    Brand_avg_price.to_excel(writer, sheet_name='Brand_Avg_Price', index=False)
```
