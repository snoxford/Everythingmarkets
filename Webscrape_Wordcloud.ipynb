```python
## This code will retrive HTML information from FT's website and will give a summary in 500 words
## A new code is added to generate a word cloud of the summary generated from FT.com

import requests
from bs4 import BeautifulSoup
from wordcloud import WordCloud 
import matplotlib.pyplot as plt


url = 'https://www.ft.com/'

response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Finding all the headings on the webpage
    paragraphs = soup.find_all('p')
    print(paragraphs)
    # Combine text from all paragraphs 
    text_content = ' '.join([para.text for para in paragraphs]) 
    
    # Create and print a summary 
    summary = ' '.join(text_content.split()[:500])
    print('\nSUMMARY\n',summary)
else:
    print(f'Failed. Status code: {response.status_code}')

# Create a word cloud for the 500 word summary created above 
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(summary) 

print(wordcloud)

# Display the word cloud using matplotlib 
plt.figure(figsize=(10, 5)) 
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis('off') 
print('\n\n')
plt.show()
```
