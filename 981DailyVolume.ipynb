```python
#Simple code to import daily trading volume for MSFT, GOOG, QQQ and SPY using yFinance

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbols
symbols = ['MSFT', 'GOOG', 'SPY', 'QQQ']

# Fetch the stock data for the last 5 years
data = {}
for symbol in symbols:
    stock_data = yf.Ticker(symbol).history(period='5y')
    data[symbol] = stock_data['Volume']

# Create a DataFrame from the fetched data
volume_df = pd.DataFrame(data)

# Plot the daily trading volume for each stock
plt.figure(figsize=(14, 7))
for symbol in symbols:
    plt.plot(volume_df.index, volume_df[symbol], label=symbol)

# Set the title and labels
plt.title('5-Year Daily Trading Volume for MSFT, GOOG, SPY, and QQQ')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()

# Show the plot
plt.show()
```
